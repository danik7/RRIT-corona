# RRITS via Fingertip Project

This notebook will try to measure RRITS values with a given fingertip video.

<b>Vocabulary</b>

R-peak and RR Intervals: In a electrocardiography graph, we call the peaks R-peaks and the distance between consecutive R-peaks RR intervals. An RRITS is an RR interval time series, which is just a list of many RR intervals over a period of time.
In addition, your pulse can be roughly calculated through the RRITS values, by dividing 60 seconds in each interval.

The RGB color model is an additive color model in which red, green, and blue light are added together in various ways to reproduce a broad array of colors. The name of the model comes from the initials of the three additive primary colors, red, green, and blue.

<b>General Info</b>

This project focus is extracting RR intervals time series by videos of a fingertip.
Individuals will record theirs own fingertip with theirs private smartphone, send that to us, 
and we will try measuring the RR values by that.
The project rely on the assumption that the fingertip colors will change according to the intervals.

<b>Work process</b>

1. Reading the video & extracting basic info
2. Measuring the total RGB values per frame
3. Comparing each color value - red, green & blue - per time
4. Trying to extract extreme points which present the R-peaks
5. Calculate RR interval and his consequences

```python
import cv2
from scipy.signal import find_peaks
from scipy.signal import argrelextrema
import matplotlib.pyplot as plt
import numpy as np
import mpld3
mpld3.enable_notebook()
import time
```

```python
# to create a new video, you will provide path only
class Video:
    # when initialising the video object, basic info is setted,
    # such as - width, height etc.
    def __init__(self, path):
        self.path = path
        self.video = cv2.VideoCapture(path)
        self.width = int(self.video.get(3))
        self.height = int(self.video.get(4))
        self.fps = self.video.get(5)
        self.frames_number = int(self.video.get(7))
        self.length = int(self.frames_number)/self.fps
        print('New video has created!')
        print(self.__dict__)
    
    # the following functions give the total RGB valus for an individual frame
    # there are few versions - some are more acurate and some are faster
    
    # FIRST version - goes over all pixels, calculating theirs own RGB value,
    # summing it up and dividing by area - width * height.
    # it's very percise but it takes 6.5 minutes for 15 second video...
    def readFrame(self):
        hasFrame, frame = self.video.read();
        width = self.width
        height = self.height
        
        if not hasFrame:
            return False, 0, 0, 0
        totalRed = 0
        totalGreen = 0
        totalBlue = 0
        for col in range(0, width):
            # adds up all of the color values for each pixel in the frame
            # returns average value per pixel
            for row in range(0, height):
                totalRed += frame[row][col][2]
                totalGreen += frame[row][col][1]
                totalBlue += frame[row][col][0]
        return True, totalRed/1.0/width/height, totalGreen/1.0/width/height, totalBlue/1.0/width/height
    
    # SECOND version - dividing the whole area to 8 pieces & calaculating 
    # the RGB values for the centered square pixels only.
    # summing it up and dividing by area - 1/8 * width * height.
    # it's not percise such as the first version but it pretty close and 
    # takes 22 seconds for 15 second video
    def readFrame2(self):
        hasFrame, frame = self.video.read();
        width = self.width
        height = self.height
        
        if not hasFrame:
            return False, 0, 0, 0
        totalRed = 0
        totalGreen = 0
        totalBlue = 0
        for col in range(int(width/8)*3, int(width/8)*5):
            # adds up all of the color values for each pixel in the frame
            # returns average value per pixel
            for row in range(int(height/8)*3, int(height/8)*5):
                totalRed += frame[row][col][2]
                totalGreen += frame[row][col][1]
                totalBlue += frame[row][col][0]
        return True, totalRed/0.125/width/height, totalGreen/0.125/width/height, totalBlue/0.125/width/height
    
    # creating an array of RGB values per frame, with the help of 'readFrame' function
    # if fast set to false, using the first 'readFrame' function
    def rgbTotalValues(self, fast=True):
        redValues = []
        greenValues = []
        blueValues= []
        
        start_time = time.time()
        print('starting frames rgb measuring...')
        count = 0
        while True:
#         while count<150:
            count += 1
            if fast: 
                hasFrame, r, g, b = self.readFrame2()
            else:
                hasFrame, r, g, b = self.readFrame()
#             hasFrame, frame = self.video.read()
            if not hasFrame:
                        break
            else:
                # the following code row is a faster and builtin function for measuring
                # total RGB value for a frame. Despite the efficiency of the function, 
                # we won't use it due to its inaccuracy
#                 b, g, r = frame[300,300]

                redValues.append(r)
                greenValues.append(g)
                blueValues.append(b)
                if count%20 == 0:
                    print('currently on frame -  {}/{}'.format(count, self.frames_number))
        
        self.total_redValues = np.array(redValues)
        self.total_greenValues = np.array(greenValues)
        self.total_blueValues = np.array(blueValues)
        
        print('\n\nTime: {}\n'.format(time.time()-start_time))
        return redValues[:-10], greenValues[:-10], blueValues[:-10]
```

```python
# function to build an array of time stamps to the frames
# with a given deltatime (usually 1/fps) and number of elements
def constructTimeList(numElements, deltaTime):
    times = [0]
    for i in range(1, numElements):
        times.append(times[i-1]+deltaTime)
    return np.array(times)
```

## Process Start

Reading the video & extracting basic info

```python
path = './amir.mp4'
video = Video(path)
```

<i>we see that the video length is about 16 seconds, with average fps - 30</i>

Measuring the total RGB values per frame

```python
print(video.rgbTotalValues())
```

```python
RV = video.total_redValues
GV = video.total_greenValues
BV = video.total_blueValues
TS = constructTimeList(len(RV), 1/video.fps)
```

```python
# finding the wanted local peaks (which represents R-peaks) indexes
def find_wanted_peaks(colorValue):
    # 1. getting an array of color values, finding all the local peaks 
    curr = colorValue
    count = 0
    last = False
    i = 2

    x = argrelextrema(curr,np.greater_equal, order=1)[0]
    last_mean = (np.mean(np.diff(x)))
    last_median = (np.median(np.diff(x)))
    
    # 2. removing the "smaller" ones each itration in the while loop
    while True:
        x = argrelextrema(curr,np.greater_equal, order=i)[0]
        mean = (np.mean(np.diff(x)))
        median = (np.median(np.diff(x)))
        if count>3:
            break
        
        # 3. waiting for the mean and the median to stabilize
        if mean==last_mean and median==last_median:
            count+=1
            last = True
        else:
            count=0
            last=False
        last_mean = mean
        last_median = median
        i+=1
        
    # 4. returning the final array of wanted peaks indexes
    return x
```

```python
# for color in ['blue', 'red', 'green']:
for color in ['red', 'green']:
    if color=='red':
        curr = RV
    elif color=='green':
        curr = GV
    elif color=='blue':
        curr = BV
    print('CURRENT COLOR - {}\n\n'.format(color))
    
    x = find_wanted_peaks(curr)
    
    # marking R-peaks
    plt.plot(x, curr[x], 'x')
    # plotting color value per frame line
    plt.plot(curr, color=color)
    plt.show()
    
    # finding the interval, by measuring difference between peaks
    final_diff = np.diff(np.array(TS)[x])
    bpm = [60/i for i in (final_diff)]

    print('The BPM among time was - {}\n'.format(str(bpm)))    
    print('the mean is {}\n'.format(np.sum(bpm)/len(bpm)))
    
    # finding standart diviation of the RR intervals
    print('STD of time intervals: {}\n\n'.format(np.std(TS[x])))
```
